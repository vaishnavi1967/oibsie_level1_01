```python
#                                           EDA on Retail Sales Data

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import warnings


warnings.filterwarnings("ignore")

df=pd.read_csv("menu.csv") 
df.head()
```

```python
df.set_index("Category",inplace=True)
```

```python

# Calculate mean

avg_total_fat=df.loc["Breakfast","Total Fat"].mean()
avg_Sugars=df.loc["Breakfast","Sugars"].mean()
avg_Carbohydrates=df.loc["Breakfast","Carbohydrates"].mean()
avg_Protein=df.loc["Breakfast","Protein"].mean()
avg_Vitamin_A_=df.loc["Breakfast","Vitamin A (% Daily Value)"].mean()

#plot 
y_Breakfast=np.array([avg_total_fat,avg_Sugars,avg_Carbohydrates,avg_Protein,avg_Vitamin_A_])
label_Breakfast=["Total Fat","Sugars","Carbohydrates","Protein","Vitamin A (% Daily Value)"]
plt.pie(y_Breakfast, labels = label_Breakfast,autopct='%1.1f%%')
plt.title("Average Nutritional Values for Breakfast Items")
plt.show() 


avg_total_fat_Smoothies_Shakes=df.loc["Smoothies & Shakes","Total Fat"].mean()
avg_Sugars_Smoothies_Shakes=df.loc["Smoothies & Shakes","Sugars"].mean()
avg_Carbohydrates_Smoothies_Shakes=df.loc["Smoothies & Shakes","Carbohydrates"].mean()
avg_Protein_Smoothies_Shakes=df.loc["Smoothies & Shakes","Protein"].mean()
avg_Vitamin_A__Smoothies_Shakes=df.loc["Smoothies & Shakes","Vitamin A (% Daily Value)"].mean()


#plot 
y_Smoothies_Shakes=np.array([avg_total_fat_Smoothies_Shakes,avg_Sugars_Smoothies_Shakes,avg_Carbohydrates_Smoothies_Shakes,avg_Protein_Smoothies_Shakes,avg_Vitamin_A__Smoothies_Shakes])
label_Smoothies_Shakes=["Total Fat","Sugars","Carbohydrates","Protein","Vitamin A (% Daily Value)"]
plt.pie(y_Smoothies_Shakes, labels = label_Smoothies_Shakes,autopct='%1.1f%%')
plt.title("Average Nutritional Values for Smoothies_Shakes")
plt.show()






```

```python
# Calculate median for Breakfast
median_total_fat = df.loc["Breakfast","Total Fat"].median()
median_Sugars = df.loc["Breakfast","Sugars"].median()
median_Carbohydrates = df.loc["Breakfast","Carbohydrates"].median()
median_Protein = df.loc["Breakfast","Protein"].median()
median_Vitamin_A_ = df.loc["Breakfast","Vitamin A (% Daily Value)"].median()

# Pie chart for Median of Breakfast Items
y_median_Breakfast = np.array([median_total_fat, median_Sugars, median_Carbohydrates, median_Protein, median_Vitamin_A_])
plt.pie(y_median_Breakfast, labels=label_Breakfast, autopct='%1.1f%%')
plt.title("Median Nutritional Values for Breakfast Items")
plt.show()


# Calculate median for Smoothies & Shakes
median_total_fat_Smoothies_Shakes = df.loc["Smoothies & Shakes","Total Fat"].median()
median_Sugars_Smoothies_Shakes = df.loc["Smoothies & Shakes","Sugars"].median()
median_Carbohydrates_Smoothies_Shakes = df.loc["Smoothies & Shakes","Carbohydrates"].median()
median_Protein_Smoothies_Shakes = df.loc["Smoothies & Shakes","Protein"].median()
median_Vitamin_A__Smoothies_Shakes = df.loc["Smoothies & Shakes","Vitamin A (% Daily Value)"].median()

# Pie chart for Median of Smoothies & Shakes
y_median_Smoothies_Shakes = np.array([median_total_fat_Smoothies_Shakes, median_Sugars_Smoothies_Shakes, median_Carbohydrates_Smoothies_Shakes, median_Protein_Smoothies_Shakes, median_Vitamin_A__Smoothies_Shakes])
plt.pie(y_median_Smoothies_Shakes, labels=label_Smoothies_Shakes, autopct='%1.1f%%')
plt.title("Median Nutritional Values for Smoothies & Shakes")
plt.show()

```

```python
# Calculate mode for Breakfast
mode_total_fat = df.loc["Breakfast","Total Fat"].mode()[0]
mode_Sugars = df.loc["Breakfast","Sugars"].mode()[0]
mode_Carbohydrates = df.loc["Breakfast","Carbohydrates"].mode()[0]
mode_Protein = df.loc["Breakfast","Protein"].mode()[0]
mode_Vitamin_A_ = df.loc["Breakfast","Vitamin A (% Daily Value)"].mode()[0]

# Pie chart for mode values of Breakfast Items
y_mode_Breakfast = np.array([mode_total_fat, mode_Sugars, mode_Carbohydrates, mode_Protein, mode_Vitamin_A_])
plt.pie(y_mode_Breakfast, labels=label_Breakfast, autopct='%1.1f%%')
plt.title("Mode Nutritional Values for Breakfast Items")
plt.show()

# Calculate mode for Smoothies & Shakes
mode_total_fat_Smoothies_Shakes = df.loc["Smoothies & Shakes","Total Fat"].mode()[0]
mode_Sugars_Smoothies_Shakes = df.loc["Smoothies & Shakes","Sugars"].mode()[0]
mode_Carbohydrates_Smoothies_Shakes = df.loc["Smoothies & Shakes","Carbohydrates"].mode()[0]
mode_Protein_Smoothies_Shakes = df.loc["Smoothies & Shakes","Protein"].mode()[0]
mode_Vitamin_A__Smoothies_Shakes = df.loc["Smoothies & Shakes","Vitamin A (% Daily Value)"].mode()[0]

# Pie chart for Mode of Smoothies & Shakes
y_mode_Smoothies_Shakes = np.array([mode_total_fat_Smoothies_Shakes, mode_Sugars_Smoothies_Shakes, mode_Carbohydrates_Smoothies_Shakes, mode_Protein_Smoothies_Shakes, mode_Vitamin_A__Smoothies_Shakes])
plt.pie(y_mode_Smoothies_Shakes, labels=label_Smoothies_Shakes, autopct='%1.1f%%')
plt.title("Mode Nutritional Values for Smoothies & Shakes")
plt.show()



```

```python
# Calculate standard deviation for BreakFast
std_total_fat = df.loc["Breakfast","Total Fat"].std()
std_Sugars = df.loc["Breakfast","Sugars"].std()
std_Carbohydrates = df.loc["Breakfast","Carbohydrates"].std()
std_Protein = df.loc["Breakfast","Protein"].std()
std_Vitamin_A_ = df.loc["Breakfast","Vitamin A (% Daily Value)"].std()

# Pie chart for standard deviation of Breakfast Items
y_std_Breakfast = np.array([std_total_fat, std_Sugars, std_Carbohydrates, std_Protein, std_Vitamin_A_])
plt.pie(y_std_Breakfast, labels=label_Breakfast, autopct='%1.1f%%')
plt.title("Standard Deviation of Nutritional Values for Breakfast Items")
plt.show()


# Calculate standard deviation for Smoothies & Shakes
std_total_fat_Smoothies_Shakes = df.loc["Smoothies & Shakes","Total Fat"].std()
std_Sugars_Smoothies_Shakes = df.loc["Smoothies & Shakes","Sugars"].std()
std_Carbohydrates_Smoothies_Shakes = df.loc["Smoothies & Shakes","Carbohydrates"].std()
std_Protein_Smoothies_Shakes = df.loc["Smoothies & Shakes","Protein"].std()
std_Vitamin_A__Smoothies_Shakes = df.loc["Smoothies & Shakes","Vitamin A (% Daily Value)"].std()

# Pie chart for Standard Deviation of Smoothies & Shakes
y_std_Smoothies_Shakes = np.array([std_total_fat_Smoothies_Shakes, std_Sugars_Smoothies_Shakes, std_Carbohydrates_Smoothies_Shakes, std_Protein_Smoothies_Shakes, std_Vitamin_A__Smoothies_Shakes])
plt.pie(y_std_Smoothies_Shakes, labels=label_Smoothies_Shakes, autopct='%1.1f%%')
plt.title("Standard Deviation of Nutritional Values for Smoothies & Shakes")
plt.show()






```

```python
# Bar graph for Breakfast Items
categories = ['Total Fat', 'Sugars', 'Carbohydrates', 'Protein', 'Vitamin A']
mean_values = [avg_total_fat, avg_Sugars, avg_Carbohydrates, avg_Protein, avg_Vitamin_A_]
median_values = [median_total_fat, median_Sugars, median_Carbohydrates, median_Protein, median_Vitamin_A_]
mode_values = [mode_total_fat, mode_Sugars, mode_Carbohydrates, mode_Protein, mode_Vitamin_A_]
std_values = [std_total_fat, std_Sugars, std_Carbohydrates, std_Protein, std_Vitamin_A_]

x = np.arange(len(categories))
width = 0.2

fig, ax = plt.subplots()
bar1 = ax.bar(x - width, mean_values, width, label='Mean')
bar2 = ax.bar(x, median_values, width, label='Median')
bar3 = ax.bar(x + width, mode_values, width, label='Mode')
bar4 = ax.bar(x + 2*width, std_values, width, label='Standard Deviation')

ax.set_xlabel('Nutritional Values')
ax.set_ylabel('Values')
ax.set_title('Nutritional Values for Breakfast Items')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```

```python
# Bar graph for Smoothies & Shakes
mean_values_smoothies = [avg_total_fat_Smoothies_Shakes, avg_Sugars_Smoothies_Shakes, avg_Carbohydrates_Smoothies_Shakes, avg_Protein_Smoothies_Shakes, avg_Vitamin_A__Smoothies_Shakes]
median_values_smoothies = [median_total_fat_Smoothies_Shakes, median_Sugars_Smoothies_Shakes, median_Carbohydrates_Smoothies_Shakes, median_Protein_Smoothies_Shakes, median_Vitamin_A__Smoothies_Shakes]
mode_values_smoothies = [mode_total_fat_Smoothies_Shakes, mode_Sugars_Smoothies_Shakes, mode_Carbohydrates_Smoothies_Shakes, mode_Protein_Smoothies_Shakes, mode_Vitamin_A__Smoothies_Shakes]
std_values_smoothies = [std_total_fat_Smoothies_Shakes, std_Sugars_Smoothies_Shakes, std_Carbohydrates_Smoothies_Shakes, std_Protein_Smoothies_Shakes, std_Vitamin_A__Smoothies_Shakes]

fig2, ax2 = plt.subplots()
bar1_smoothies = ax2.bar(x - width, mean_values_smoothies, width, label='Mean')
bar2_smoothies = ax2.bar(x, median_values_smoothies, width, label='Median')
bar3_smoothies = ax2.bar(x + width, mode_values_smoothies, width, label='Mode')
bar4_smoothies = ax2.bar(x + 2*width, std_values_smoothies, width, label='Standard Deviation')

ax2.set_xlabel('Nutritional Values')
ax2.set_ylabel('Values')
ax2.set_title('Nutritional Values for Smoothies & Shakes')
ax2.set_xticks(x)
ax2.set_xticklabels(categories)
ax2.legend()

plt.show()
```

```python
df_2=pd.read_csv("retail_sales_dataset.csv") 
df_2
```

```python
# Convert 'Date' column to datetime
df_2['Date'] = pd.to_datetime(df_2['Date'])

# Set 'Date' column as the index
df_2.set_index('Date', inplace=True)

# Resample data to monthly frequency and sum the total amount for each month
monthly_sales = df_2['Total Amount'].resample('M').sum()

# Plotting sales trends over time
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-')
plt.title('Monthly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Customer demographics analysis
customer_demographics = df_2.groupby('Gender')['Age'].describe()

# Product purchasing behavior analysis
# Product purchasing behavior analysis
product_purchasing = df_2.groupby('Product Category')[['Quantity', 'Total Amount']].sum()


# Visualizing customer demographics
customer_demographics.plot(kind='bar', figsize=(8, 6), rot=0)
plt.title('Customer Demographics')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Visualizing product purchasing behavior
product_purchasing.plot(kind='bar', figsize=(10, 6))
plt.title('Product Purchasing Behavior')
plt.xlabel('Product Category')
plt.ylabel('Quantity/Total Amount')
plt.show()
```
